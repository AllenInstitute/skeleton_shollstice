```python
import os
import pandas as pd

path_to_swc_directors = "./test_swc_files/"
all_sholl_results = pd.DataFrame()
for fn in os.listdir(path_to_swc_directors):
    if fn.endswith(".swc"):
        swc_path = os.path.join(path_to_swc_directors,fn)
        sholler = Shollster(swc_path)
        sholl_res = sholler.run_sholl(sholl_method="intersections",compatment_types=[3])
        sholl_res = sholl_res.T
        sholl_res['swc_path'] = swc_path
        
        all_sholl_results = pd.concat([all_sholl_results,sholl_res])

all_sholl_results = all_sholl_results.fillna(0)
all_sholl_results = all_sholl_results.set_index('swc_path')
```
